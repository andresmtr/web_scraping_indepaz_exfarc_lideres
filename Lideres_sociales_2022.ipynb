```python
# importar librerias

import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import csv
from pandas import ExcelWriter
```

```python
# extración población por País

website = requests.get('https://indepaz.org.co/lideres-sociales-defensores-de-dd-hh-y-firmantes-de-acuerdo-asesinados-en-2022/')
```

# Lideres sociales asesinados

```python
# Codigo

soup = BeautifulSoup(website.text, 'html.parser')

first_table = soup.select('.has-background')[0]

first_table
```

```python
table_row = first_table.select('tr')[1:]

table_row
```

```python
csv_data = [['No','Nombre','Fecha','Municipio','Departamento','Calidad']]
```

```python
for row in table_row[0:]:
    
    
    table_data = row.select('td')
    No = table_data[0].text
    Nombre = table_data[1].text
    Fecha = table_data[2].text
    Municipio = table_data[3].text
    Departamento = table_data[4].text
    Calidad = table_data[5].text
    
    csv_data.append([No, Nombre, Fecha, Municipio, Departamento, Calidad])
```

```python
csv_data
```

```python
lideres = pd.DataFrame(csv_data, columns=csv_data[0])
lideres=lideres.drop(lideres.index[[0]])
lideres
```

```python
# Depuración departamentos

lideres['Departamento'].replace(
    to_replace=['BOGOTA'],
    value='BOGOTÁ',
    inplace=True
)


lideres['Departamento'].replace(
    to_replace=['BOLIVAR'],
    value='BOLÍVAR',
    inplace=True
)

lideres['Departamento'].replace(
    to_replace=['CALDAS'],
    value='CÁLDAS',
    inplace=True
)

lideres['Departamento'].replace(
    to_replace=['CAQUETA'],
    value='CAQUETÁ',
    inplace=True
)

lideres['Departamento'].replace(
    to_replace=['CHOCO'],
    value='CHOCÓ',
    inplace=True
)

lideres['Departamento'].replace(
    to_replace=['CORDOBA'],
    value='CÓRDOBA',
    inplace=True
)

lideres['Departamento'].replace(
    to_replace=['GUAJIRA'],
    value='LA GUAJIRA',
    inplace=True
)

lideres['Departamento'].replace(
    to_replace=['GUAJIRA'],
    value='LA GUAJIRA',
    inplace=True
)

lideres['Departamento'].replace(
    to_replace=['VALLE DELCAUCA'],
    value='VALLE DEL CAUCA',
    inplace=True
)
```

```python
# Depuración municipios

# Depuración departamentos

lideres['Municipio'].replace(
    to_replace=['BOGOTA'],
    value='BOGOTÁ',
    inplace=True
)

lideres['Municipio'].replace(
    to_replace=['ALTO BAUDO'],
    value='ALTO BAUDÓ',
    inplace=True
)

lideres['Municipio'].replace(
    to_replace=['CAJIBIO'],
    value='CAJIBÍO',
    inplace=True
)


lideres['Municipio'].replace(
    to_replace=['CI´ÉNAGA'],
    value='CIÉNAGA',
    inplace=True
)
```

```python
lideres['Departamento'] = lideres['Departamento'].str.strip()

lideres['Departamento'] = lideres['Departamento'].str.upper()

lideres['Municipio'] = lideres['Municipio'].str.strip()

lideres['Municipio'] = lideres['Municipio'].str.upper()
```

```python


departamentos = lideres.groupby(['Departamento']).count().reset_index()

departamentos = pd.concat([departamentos['Departamento'], departamentos['No']],axis=1)

departamentos.columns = ['Departamento', 'No']

departamentos
```

```python
departamentos_municipio = lideres.groupby(['Departamento', 'Municipio']).count().reset_index()

departamentos_municipio = pd.concat([departamentos_municipio['Departamento'], departamentos_municipio['Municipio'], departamentos_municipio['No']],axis=1)

departamentos_municipio.columns = ['Departamento', 'Municipio','No']

departamentos_municipio
```

```python
#Lideres s

with ExcelWriter('Lideres_sociales_2022.xlsx') as writer:
    lideres.to_excel(writer, 'Lista_completa')
    departamentos.to_excel(writer, 'total departamento')
    departamentos_municipio.to_excel(writer, 'total municipio')
```
